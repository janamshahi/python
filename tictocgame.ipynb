```python
import tkinter as tk
from tkinter import messagebox

root = tk.Tk()
root.title("Tic-Tac-Toe AI")

# Game variables
board = [""] * 9
buttons = []
human = "X"
ai = "O"

# Check winner
def check_winner(b, player):
    win = [
        [0,1,2],[3,4,5],[6,7,8],
        [0,3,6],[1,4,7],[2,5,8],
        [0,4,8],[2,4,6]
    ]
    for c in win:
        if all(b[i] == player for i in c):
            return True
    return False

def is_draw(b):
    return "" not in b

# Minimax AI
def minimax(b, is_max):
    if check_winner(b, ai):
        return 1
    if check_winner(b, human):
        return -1
    if is_draw(b):
        return 0

    if is_max:
        best = -100
        for i in range(9):
            if b[i] == "":
                b[i] = ai
                score = minimax(b, False)
                b[i] = ""
                best = max(best, score)
        return best
    else:
        best = 100
        for i in range(9):
            if b[i] == "":
                b[i] = human
                score = minimax(b, True)
                b[i] = ""
                best = min(best, score)
        return best

# Find best move
def best_move():
    best_score = -100
    move = -1
    for i in range(9):
        if board[i] == "":
            board[i] = ai
            score = minimax(board, False)
            board[i] = ""
            if score > best_score:
                best_score = score
                move = i
    return move

# Handle player click
def on_click(i):
    if board[i] == "":
        board[i] = human
        buttons[i].config(text=human)

        if check_winner(board, human):
            messagebox.showinfo("Game Over", "You win! 🎉")
            reset_game()
            return
        
        if is_draw(board):
            messagebox.showinfo("Game Over", "Draw 🤝")
            reset_game()
            return

        # AI move
        ai_move = best_move()
        if ai_move != -1:
            board[ai_move] = ai
            buttons[ai_move].config(text=ai)

        if check_winner(board, ai):
            messagebox.showinfo("Game Over", "Computer wins 🤖")
            reset_game()
            return
        
        if is_draw(board):
            messagebox.showinfo("Game Over", "Draw 🤝")
            reset_game()

# Reset game
def reset_game():
    global board
    board = [""] * 9
    for btn in buttons:
        btn.config(text="")

# Create buttons
for i in range(9):
    btn = tk.Button(root, text="", font=("Arial", 24),
                    width=5, height=2,
                    command=lambda i=i: on_click(i))
    btn.grid(row=i//3, column=i%3)
    buttons.append(btn)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

root = tk.Tk()
root.title("Tic-Tac-Toe AI")
root.configure(bg="#1e1e2f")

board = [""] * 9
buttons = []
human = "X"
ai = "O"

status = tk.Label(root, text="Your Turn (X)", font=("Arial", 14),
                  bg="#1e1e2f", fg="white")
status.grid(row=3, column=0, columnspan=3, pady=10)

# ---------------- AI LOGIC ----------------
def check_winner(b, player):
    wins = [
        [0,1,2],[3,4,5],[6,7,8],
        [0,3,6],[1,4,7],[2,5,8],
        [0,4,8],[2,4,6]
    ]
    for w in wins:
        if all(b[i] == player for i in w):
            return w
    return None

def is_draw(b):
    return "" not in b

def minimax(b, is_max):
    if check_winner(b, ai):
        return 1
    if check_winner(b, human):
        return -1
    if is_draw(b):
        return 0

    if is_max:
        best = -100
        for i in range(9):
            if b[i] == "":
                b[i] = ai
                score = minimax(b, False)
                b[i] = ""
                best = max(best, score)
        return best
    else:
        best = 100
        for i in range(9):
            if b[i] == "":
                b[i] = human
                score = minimax(b, True)
                b[i] = ""
                best = min(best, score)
        return best

def best_move():
    best_score = -100
    move = -1
    for i in range(9):
        if board[i] == "":
            board[i] = ai
            score = minimax(board, False)
            board[i] = ""
            if score > best_score:
                best_score = score
                move = i
    return move

# ---------------- UI EFFECTS ----------------
def animate_click(btn):
    btn.config(bg="#00adb5")
    root.after(100, lambda: btn.config(bg="#393e46"))

def highlight_winner(cells):
    for i in cells:
        buttons[i].config(bg="#00ff99")

# ---------------- GAME ----------------
def on_click(i):
    if board[i] == "":
        board[i] = human
        buttons[i].config(text=human, fg="#00adb5")
        animate_click(buttons[i])

        win = check_winner(board, human)
        if win:
            highlight_winner(win)
            status.config(text="You Win! 🎉")
            root.after(1500, reset_game)
            return

        if is_draw(board):
            status.config(text="Draw 🤝")
            root.after(1500, reset_game)
            return

        status.config(text="Computer Thinking... 🤖")
        root.after(500, ai_turn)

def ai_turn():
    move = best_move()
    if move != -1:
        board[move] = ai
        buttons[move].config(text=ai, fg="#ff2e63")
        animate_click(buttons[move])

    win = check_winner(board, ai)
    if win:
        highlight_winner(win)
        status.config(text="Computer Wins 🤖")
        root.after(1500, reset_game)
        return

    if is_draw(board):
        status.config(text="Draw 🤝")
        root.after(1500, reset_game)
        return

    status.config(text="Your Turn (X)")

def reset_game():
    global board
    board = [""] * 9
    for btn in buttons:
        btn.config(text="", bg="#393e46")
    status.config(text="Your Turn (X)")

# ---------------- BUTTON GRID ----------------
for i in range(9):
    btn = tk.Button(root, text="", font=("Arial", 24),
                    width=5, height=2,
                    bg="#393e46", fg="white",
                    activebackground="#00adb5",
                    relief="flat",
                    command=lambda i=i: on_click(i))

    # Hover effect
    btn.bind("<Enter>", lambda e, b=btn: b.config(bg="#50555c"))
    btn.bind("<Leave>", lambda e, b=btn: b.config(bg="#393e46"))

    btn.grid(row=i//3, column=i%3, padx=5, pady=5)
    buttons.append(btn)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

# Create main window
root = tk.Tk()
root.title("Tic-Tac-Toe")

# Game variables
current_player = "X"
board = [""] * 9
buttons = []

# Check winner
def check_winner(player):
    win_conditions = [
        [0,1,2], [3,4,5], [6,7,8],
        [0,3,6], [1,4,7], [2,5,8],
        [0,4,8], [2,4,6]
    ]
    for condition in win_conditions:
        if all(board[i] == player for i in condition):
            return True
    return False

# Check draw
def is_draw():
    return "" not in board

# Button click event
def on_click(i):
    global current_player
    
    if board[i] == "":
        board[i] = current_player
        buttons[i].config(text=current_player)
        
        if check_winner(current_player):
            messagebox.showinfo("Game Over", f"Player {current_player} wins!")
            reset_game()
            return
        
        if is_draw():
            messagebox.showinfo("Game Over", "It's a draw!")
            reset_game()
            return
        
        # Switch player
        current_player = "O" if current_player == "X" else "X"

# Reset game
def reset_game():
    global current_player, board
    current_player = "X"
    board = [""] * 9
    for btn in buttons:
        btn.config(text="")

# Create buttons
for i in range(9):
    btn = tk.Button(root, text="", font=("Arial", 24), width=5, height=2,
                    command=lambda i=i: on_click(i))
    btn.grid(row=i//3, column=i%3)
    buttons.append(btn)

# Run GUI
root.mainloop()
```
